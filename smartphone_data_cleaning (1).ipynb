## Data Assessing

18. **os** - sometimes contains os version name like lollipop `consistency`
19. missing values in camera, card and os `completeness`
20. datatype  of price and rating is incorrect `validity`



### Tidiness Issues

1. **sim** - can be split into 3 cols has_5g, has_NFC, has_IR_Blaster
2. **ram** - can be split into 2 cols RAM and ROM
3. **processor** - can be split into processor name, cores and cpu speed.
4. **battery** - can be split into battery capacity, fast_charging_available
5. **display** - can be split into size, resolution_width, resolution_height and frequency
6. **camera** - can be split into front and rear camera
7. **card** - can be split into supported, extended_upto

```python
# camera counter function
def count_cam(text):
  if "Quad" in text:
    return 4
  elif "Triple" in text:
    return 3
  elif "Dual" in text:
    return 2
  else:
    return 1
```

```python
# extract no of back camera
n_rear_cam = df1.camera.str.split("&").str.get(0).apply(count_cam)
```

```python
# insert new col n_rear_cam
df1.insert(20,"n_rear_cam", n_rear_cam)
#drop camera col
df1.drop(columns="camera", inplace=True)
```

```python
#i think we should drop card column because this info not much required and not useful
df1.drop(columns = "card", inplace = True)
```

```python
# Now our cleaning is Done
# Recheck everything one by one
# then start eda

# checking null values
# rating, coretype,speed has little null values
#fill with mean
df1.rating.fillna(df1.rating.mean().round(),inplace=True)
```

```python
# handle core_type missing data
# fill with mode value
df1.core_type.fillna("octa core",inplace = True )
```

```python
# handle core_type missing data
# fill with mean value
df1.speed.fillna(df1.speed.mean().round(),inplace=True)
```
